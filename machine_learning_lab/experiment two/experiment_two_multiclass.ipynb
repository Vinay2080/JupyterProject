```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# Example with three classes
y_true = [0, 1, 2, 0, 1, 2, 0, 1, 2, 1, 0, 2, 1, 0, 2]
y_pred = [0, 2, 1, 0, 1, 2, 1, 1, 2, 0, 0, 2, 1, 0, 1]
```

```python
confusion_matrix(y_true, y_pred)
```

```python
accuracy_score(y_true, y_pred)
```

Precision is calculated for each class by treating that class as the positive class and the others as negative. The average can be calculated in different ways:
- **Macro Average**: Calculate the precision for each class and then take the average. This treats all classes equally, regardless of their support (number of true instances for each class).
- **Micro Average**: Calculate the total true positives, false positives, and false negatives across all classes and then compute the precision. This gives more weight to classes with more instances.
- **Weighted Average**: Calculate the precision for each class and then take the average, weighted by the number of true instances for each class. This accounts for class imbalance by giving more weight to classes with more instances.

```python
precision_score(y_true, y_pred, average='macro')
```

```python
precision_score(y_true, y_pred, average='micro')
```

```python
precision_score(y_true, y_pred, average='weighted')
```

```python
recall_score(y_true, y_pred, average='macro')
```

```python
recall_score(y_true, y_pred, average='micro')
```

```python
recall_score(y_true, y_pred, average='weighted')
```

```python
f1_score(y_true, y_pred, average='macro')
```

```python
f1_score(y_true, y_pred, average='micro')
```

```python
f1_score(y_true, y_pred, average='weighted')
```

```python
cm = confusion_matrix(y_true, y_pred)

# For Class 0 (treating class 0 as positive, classes 1&2 as negative)
tn_class0 = cm[1, 1] + cm[1, 2] + cm[2, 1] + cm[2, 2]  # True negatives for class 0
fp_class0 = cm[0, 1] + cm[0, 2]  # False positives for class 0
fn_class0 = cm[1, 0] + cm[2, 0]  # False negatives for class 0
tp_class0 = cm[0, 0]  # True positives for class 0

# For Class 1 (treating class 1 as positive, classes 0&2 as negative)
tn_class1 = cm[0, 0] + cm[0, 2] + cm[2, 0] + cm[2, 2]  # True negatives for class 1
fp_class1 = cm[1, 0] + cm[1, 2]  # False positives for class 1
fn_class1 = cm[0, 1] + cm[2, 1]  # False negatives for class 1
tp_class1 = cm[1, 1]  # True positives for class 1

# For Class 2 (treating class 2 as positive, classes 0&1 as negative)
tn_class2 = cm[0, 0] + cm[0, 1] + cm[1, 0] + cm[1, 1]  # True negatives for class 2
fp_class2 = cm[2, 0] + cm[2, 1]  # False positives for class 2
fn_class2 = cm[0, 2] + cm[1, 2]  # False negatives for class 2
tp_class2 = cm[2, 2]  # True positives for class 2

print(f"Class 0: TN={tn_class0}, FP={fp_class0}, FN={fn_class0}, TP={tp_class0}")
print(f"Class 1: TN={tn_class1}, FP={fp_class1}, FN={fn_class1}, TP={tp_class1}")
print(f"Class 2: TN={tn_class2}, FP={fp_class2}, FN={fn_class2}, TP={tp_class2}")
```
